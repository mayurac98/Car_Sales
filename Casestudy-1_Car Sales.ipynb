# 3.Plot a bar graph of male vs female buyers participated in the sales

```python
plt.figure(figsize=(7,5))
data1['Buyer Gender'].value_counts().plot(kind='bar',rot=0,color=['violet','blue'])
plt.xlabel("Gender")
plt.ylabel("Count of People")
plt.title("Participation in sales");
```
